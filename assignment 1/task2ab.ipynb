```python
import matplotlib.pyplot as plt
import pathlib
import numpy as np
from utils import read_im, save_im
output_dir = pathlib.Path("image_solutions")
output_dir.mkdir(exist_ok=True)
```

```python
im = read_im(pathlib.Path("images", "duck.jpeg"))
plt.imshow(im)
```

```python
def greyscale(im):
    """ Converts an RGB image to greyscale
    
    Args:
        im ([type]): [np.array of shape [H, W, 3]]
    
    Returns:
        im ([type]): [np.array of shape [H, W]]
    """
    H, W, D = im.shape
    grey = np.zeros((H, W))
    for h in range(H):
        for w in range(W):
             grey[h][w] = 0.212 * im[h][w][0] + 0.7152 * im[h][w][1] + 0.0722 * im[h][w][2]

    return grey
```

```python
im_greyscale = greyscale(im)
save_im(output_dir.joinpath("duck_greyscale.jpeg"), im_greyscale, cmap="gray")
plt.imshow(im_greyscale, cmap="gray")
```

## Task 2b

```python
def inverse(im):
    """ Finds the inverse of the greyscale image
    
    Args:
        im ([type]): [np.array of shape [H, W]]
    
    Returns:
        im ([type]): [np.array of shape [H, W]]
    """    
    H, W = im.shape

    for h in range(H):
        for w in range(W):
            im[h][w] = 255-im[h][w]

    return im
```

```python
im_inverse = inverse(im_greyscale)
save_im(output_dir.joinpath("duck_greyscale_inverse.jpeg"), im_inverse, cmap="gray")
plt.imshow(im_inverse, cmap="gray")
```
